### Step-1: Import Libraries

```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

### Step-2: Load and Explore Data

```python
# Load shape file
File_path = "C:/Users/USER/Desktop/JUPYTER/DATA/ne_10m_populated_places.shp"
data = gpd.read_file(File_path)
# Explore data
print(data.head(12))
print (data.info())
```

### Step-3 Filtering Data

```python
filterdata = data[data['SOV0NAME'] == 'India']
print (filterdata.head())
```

### Step-3: Data Visualization

```python
## Plot the geospatial data
filterdata.plot()
plt.title('India Population data')
plt.grid()
plt.show()
```
