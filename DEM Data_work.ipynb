```python
import rasterio
from rasterio.plot import show
import matplotlib.pyplot as plt
```

```python
dem_file = "C:/Users/USER/Desktop/JUPYTER/DATA/DEM Data/West_Bengal_DEM.tif"
```

```python
# open the DEM File
dem_dataset = rasterio.open(dem_file)
```

```python
# read the DEM Data
dem_array = dem_dataset.read(1)
```

```python
# print some iformation about DEM
print(f"DEM Shape: {dem_array.shape}")
print(f"SEM Metadata: {dem_dataset.meta}")
```

## Visualize DEM Data

```python
# plot the DEM data using mapplotlib
plt.figure(figsize=(10,10))
show(dem_dataset, cmap='terrain')
#plt.title('Digital Elevation Model (DEM)')
plt.show()
```
